```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10) # 0 to before 10 generates integers
# for all these x's we want to generate a y
Y = (X - 5 ) **2 +3 #we are taking any sample function
print(X, Y)
```

```python
# Visualising curve
plt.style.use('seaborn')
plt.plot(X,Y)
plt.ylabel('Y')
plt.xlabel('X')
plt.show


```

```python
#Gradient Descent
#at what value of X value of Y is minimum, in our case For X =5 we get the minimum value
# and the minimum value is 3

#now we need to apply grad descent and find 3

x = 9 # starting with current random value, also starting from either left or right side yields the same result
# now we utilise our update
lr = 0.1  # learning rate, if you put too big of a LR it wont converge, and if its too
#small it will take many steps to reach the answer
# gradient is dy/dx or our slope 'm' for our case x-5**2 +3 grad = 2(x-5) + 0
plt.plot(X,Y)
for i in range(30):
  gradient = 2*(x-5)
  x = x-lr*gradient
  y = (x - 5 )**2 +3
  plt.scatter(x,y)
  print(x)
plt.show()
# we notice that our answer converges at x value of 5, since grad is decreasing as we move down
#
```
